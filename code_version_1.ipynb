```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load dataset from CSV file
data = pd.read_csv('Bangalore.csv') 
```

```python
# Initial Data Exploration
print("First few rows of the dataset:")
print(data.head())

print("\nData Summary:")
print(data.info())
```

```python
# Check for missing values
print("\nMissing Values in Each Column:")
print(data.isnull().sum())
```

```python
# Summary statistics for further insights
print("Summary Statistics:")
print(data.describe())
```

```python
#  Data Analysis

# 1. Distribution of Prices
plt.figure(figsize=(10, 6))
sns.histplot(data['Price'], kde=True)
plt.title("Distribution of House Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# 2. Price vs Area
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Area', y='Price', data=data)
plt.title("Price vs Area")
plt.xlabel("Area (sq. ft.)")
plt.ylabel("Price")
plt.show()
```

```python
# 3. Price by No. of Bedrooms
plt.figure(figsize=(10, 6))
sns.boxplot(x='No. of Bedrooms', y='Price', data=data)
plt.title("Price by Number of Bedrooms")
plt.xlabel("Number of Bedrooms")
plt.ylabel("Price")
plt.show()
```

```python
# Re-define the amenity columns list
amenity_columns = [
    'MaintenanceStaff', 'Gymnasium', 'SwimmingPool', 'LandscapedGardens', 'JoggingTrack', 
    'RainWaterHarvesting', 'IndoorGames', 'ShoppingMall', 'Intercom', 'SportsFacility', 
    'ATM', 'ClubHouse', 'School', '24X7Security', 'PowerBackup', 'CarParking', 
    'StaffQuarter', 'Cafeteria', 'MultipurposeRoom', 'Hospital', 'WashingMachine', 
    'Gasconnection', 'AC', 'Wifi', "Children'splayarea", 'LiftAvailable', 'BED', 
    'VaastuCompliant', 'Microwave', 'GolfCourse', 'TV', 'DiningTable', 'Sofa', 
    'Wardrobe', 'Stadium'
]

# Ensure that AmenityCount is calculated
if 'AmenityCount' not in data.columns:
    data['AmenityCount'] = data[amenity_columns].sum(axis=1)

# Plot Price by Amenity Count
plt.figure(figsize=(10, 6))
sns.scatterplot(x='AmenityCount', y='Price', data=data)
plt.title("Price by Amenity Count")
plt.xlabel("Amenity Count")
plt.ylabel("Price")
plt.show()
```
